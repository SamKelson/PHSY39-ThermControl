```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataY = np.array([50.0, 45.7, 44.7, 44.0, 42.6, 42.1, 39.2, 36.8, 36.2])
dataX = np.array([0, 0.5, 1.0, 1.5, 2.0, 2.5, 4.5, 6.5, 8.5])

#new x's
#dataX = np.array([0, 0.5, 1.0, 2.5, 3.5, 4.5])
```

```python
fig, ax = plt.subplots()
ax.plot(dataX,dataY)
q=0.055
ax.plot(dataX,50*np.exp(-1*q*(dataX)),label="1/q="+str(1/q))
ax.set(xlabel='Dist from TEC (inch)', ylabel='Temp (C)', title='Temperature vs. Distance from TEC')
ax.legend()
```
